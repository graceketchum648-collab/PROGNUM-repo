```python
from math import atan2, degrees
x = float(input("Enter a value for x: "))
y = float(input("Enter a value for y: "))

a = atan2(y,x) #the angle
ad = degrees (a) #the angle in degrees

if -pi <a< pi/2 :
    print (f"The angle is {ad}, and {x,y} belongs to the first quadrant.")
elif pi/2 <a< pi :
    print (f"The angle is {ad}, and {x,y} belongs to the second quadrant.")
elif pi <a< -pi/2 :
    print (f"The angle is {ad}, and {x,y} belongs to the third quadrant.")
else :
    print (f"The angle is {ad}, and {x,y} belongs to the fourth quadrant.")
```
