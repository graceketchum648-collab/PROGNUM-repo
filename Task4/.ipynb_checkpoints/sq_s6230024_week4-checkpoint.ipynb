```python
x = [1,2,3,10] #creating both lists
y = [10, 20, 30, 40]


z1=[x+y for x, y in zip(x, y)]#using list comprehension and making the code one line
print(z1)

def addxy(x, y): #defining a function to use map()
    z=x+y
    return z

sumxy=map(addxy, x, y) #entering the function as well as the two variables
z2=list(sumxy)#turning it into a list

print(z2)
```

```python
x = [1,2,3,10] #creating both lists
y = [10, 20, 30, 40]

l=[]
zi=zip(x,y)
for x,y in zi:
    z=x+y
    l.append(z)
    
print(l)
```

```python
import numpy as np
import scipy
from matplotlib import pyplot as plt

lengths=np.random.normal(185, 8, 100000) #drawing random values from a normal distribution with a mean of 185 and a stdv of 8

mcalc=np.mean(lengths) #calculating the mean using numpy function
stdvcalc=np.std(lengths) #calculating the standard deviation using numpy function
print(f"The mean is {mcalc} and the standard deviation is {stdvcalc}")

h1=float(input("Choose a value for h1:"))
cdf1=scipy.stats.norm.cdf(h1, 185, 8) #finding the probability that any value of x is smaller or equal to h
print(f"The probability of any value being smaller or equal to your chosen height is {cdf1}")

cdfa=scipy.stats.norm.cdf(185+8, 185, 8)#defining the probability that a height lies within the range of 0 to mean+stdv
cdfb=scipy.stats.norm.cdf(185-8, 185, 8)#defining the probability that a height lies within the range of 0 to mean-stdv

cdf2=cdfa-cdfb#subtracting the two to get the probability that the height is within mean +- stdv
print(f"The probability of a height being located within 1 standard deviation range is {cdf2}")
      
plt.hist(lengths, bins=40, label=f"mean={mcalc}, standard deviation={stdvcalc}") #plotting the historigram
plt.axvline(185, color="r", label="The sample mean") #plotting the vertical line
plt.axvspan(185-8, 185+8, color='g', alpha=0.2) #plotting the shaded range
plt.legend() #plotting the legends
plt.show
```

```python
import math
import numpy as np
from matplotlib import pyplot as plt

angle_degs = 30  #Rotation angle
theta = math.radians(angle_degs)

rotor = math.cos(theta) + 1j*math.sin(theta)#defining the rotor vector


x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

square=np.array([1+1j, 3+1j, 3+3j, 1+3j, 1+3j, 1+1j]) #turning the arrays into complex

squarer=rotor*square #rotating the square

xr=squarer.real #the x's are the real component of the complex vectors
yr=squarer.imag #the y's are the imaginary component of the complex vectors

plt.plot(x, y)
plt.plot(xr, yr, color='r')
plt.show
```

```python
import numpy as np

def funcfilter(func, seq):#defining the filter functiom
    l=[] #creating a list to append the values to
    for i in range(len(seq)):
        if seq[i]>0: #usingf the indices to use the values themselves
            res=func(seq[i])
            l.append(res)#adding the results to the list
    return (l)


sq=np.array([-100, -10, 0, 10, 100]) #test n1
result = funcfilter(np.log10, sq)
print("Result log10:", result)


sq2=np.asarray([100]) #test n2
res2= funcfilter(np.log10, sq2)
print("Result2:", res2)
```

```python
from astropy.cosmology import Planck18
from scipy import integrate

z=2 #defining the redshift

H0 = Planck18.H0.value #defining the constants
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0


def t(z): #defining the function
    y=1/((1+z)*(H0*(((Om0*((1+z)**3))+Ode0)**(1/2))))
    return y

intt=integrate.quad(t, 0, z) #using the quad to indtegrate from 0 to z 

result_gyr=intt[0]*977.8 #converting it to gyr using a factor converting Mpc to km and s to years

print(f"The lookback time for z=2 is {result_gyr:.2f} Gyrs")
```

```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt

X=np.random.normal(0, 1, 100000) #generating the values
Y=np.random.normal(2, 1, 100000)

x_values=np.linspace(-6, 8, 1000)
 #getting x values that match the data provided

plt.hist(X, bins=30, alpha=0.5, label='X', density=True, color='b', edgecolor='black') #plotting teh first histogram
plt.hist(Y, bins=30, alpha=0.5, label='Y', density=True, color='r', edgecolor='black') #plotting the first histogram
plt.plot(x_values, norm.pdf(x_values, 0, 1), label='X(PDF)', color='black', linewidth=4, linestyle=':') # PDFs using scipy
plt.plot(x_values, norm.pdf(x_values, 2, 1),label='Y(PDF)', color='black', linewidth=4, linestyle='--')

plt.annotate(r'$\mathcal{N}(0, 1)$', #fancy n
             xy=(-2, 0.15),     #point of teh arrow
             xytext=(-5, 0.15),  #placement of teh text
             fontsize=14,        
             color='blue',        
             va='center',         
             arrowprops=dict(width=8, headwidth=12, color='blue'))

plt.annotate(r'$\mathcal{N}(2, 1)$', 
             xy=(4, 0.15),
             xytext=(6, 0.15),
             fontsize=14, 
             color='red', 
             va='center', 
             ha='center', 
             arrowprops=dict(width=8, headwidth=12, color='red'))
plt.xlabel('Values', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.title('1D Histogram + PDF', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(range(-6, 9))
plt.xlim(-6, 8)
plt.ylim(0, 0.43)
plt.legend()
plt.show
```
