```python
import numpy as np
from numpy import sin, cos, exp, pi
import scipy


a=eval(input("Input a lower bound:")) #letting teh user choose the lower bound
b=eval(input("Input an upper bound:")) #letting the user choose the upper bound
n=1000 #making them choose a number of data points 
fofx= input("Input a function f(x) using python syntax:") #letting them pick a function
x=np.random.uniform(a, b, n) #choosing x values from a random uniform distribution

f=lambda x: eval(fofx)

intt = scipy.integrate.quad(f, a, b) #using eval to create a list of all of the values of the function
inttmc=((b-a)/n)*sum(f(x))

print(intt)
print(inttmc)
```
