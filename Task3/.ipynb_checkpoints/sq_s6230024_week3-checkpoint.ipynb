```python
from math import sqrt
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

v_escl=[] #creating an empty list to add the escape values

number_of_objects = len(masses)
indxlist = range(number_of_objects)

for i in indxlist: #make a loop to calculate the esc velocities withh the masses and radi of the same index
    m=masses[i]
    r=radii[i]
    v_esc= sqrt((2*G*m)/r)
    v_escl.append(v_esc) #adding the escape velocities to the list created before
    print(f" The escape velocity is {v_esc}, for a mass {masses[i]:6.2f}" )

dictionary=dict(zip(names, v_escl)) #creating a dictiary based off of both lists
print(dictionary)
```

```python
import math
h=float(input("Input a value for h:"))
r=float(input("Input a value for r:")) 

def cylinder(h, r):
    A=(2*math.pi*(r**2))+(2*math.pi*r*h) #formula for the are of a cylinder
    V=math.pi*(r**2)*h #formula for the volume of a cylinder
    return (A,V)

cyl= cylinder(h, r)
print(f"Given the area and radius, the area and volume of the cylinder are {cyl}")

Acyl= cyl[0]
Vcyl= cyl[1]

rat= Acyl/Vcyl
print(f"The ratio is {rat}")
```

```python
import numpy as np

x=np.array([1, 2, 3, 4]) #any input array

def stdev(x): #x is an array
    N=x.size #defining N as the number of variables in the array
    mu= np.sum(x)/N #calculating the mean
    sumsquared= (np.sum((x-mu)**2))/N #calculating what is under the root
    sigma= np.sqrt(sumsquared)
    return sigma

test= stdev(x)
print(f"The created function gives a std of {test}")
deffunc= np.std(x)
print(f"The numpy function gives a std of {deffunc}")


def stdev2(x):
    sigma=np.sqrt((np.sum((x-(np.sum(x)/x.size))**2))/x.size)
    return sigma
test2=stdev2(x)
print(f"The created one liner gives a std of {test2}")
```

```python
import numpy as np

x=np.array([18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19])
y=np.array([2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 1410, 2700, 5292, 2550, 5253, 2419, 2158])

xbar=np.sum(x)/(x.size) #the average of x
ybar=np.sum(y)/(y.size) #the average of y

b=(np.sum((x-xbar)*(y-ybar)))/(np.sum((x-xbar)**2))
a= ybar-(b*xbar)

print(f"The value for a is {a}, and the value for b is {b}")
```

```python
import matplotlib.pyplot as plt
import math

x_values=range(-5, 6) #starts at -4 in order to exclude -5
def shape(x):
    shape=math.cosh(x)
    return shape
y_values= [shape(x) for x in x_values]

plt.plot(x_values, y_values, label="Catenary")
plt.title("Catenary")
plt.grid()
plt.ylabel("Hyperbolic cosine", fontsize=16)
plt.xlabel("Values of x", fontsize=16)
plt.xticks(x_values)
plt.yticks(range(-1, 75, 5))
plt.legend()
plt.show()
```
