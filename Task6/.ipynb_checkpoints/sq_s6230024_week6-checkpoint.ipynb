```python
import numpy as np
from astropy.io import fits
from scipy import stats
import matplotlib.pyplot as plt



m101= fits.open('m101.fits')
m101.info() #there are 2 HDU's, primary and photometric caltabele

#The primary hdu has the image data (as it is 530x530), it is index [0]

#the number type is int16

hdr=m101[0].header
for key in hdr[:5]:
    print(key,hdr[key]) #listing the first five headers

#the bitpix is 16, correspodnding w the previous data

dat = m101[0].data
flatdat = dat.flatten()

print(f"shape={dat.shape}") #This gives (530x530) which corresponds to what was given with info()

#this corresponds to the naxis values


print(f"Data type: {dat.dtype}") #gives >i2, meaning 2 byte integer, 2x8, thus corresponding with int16

fmin=np.min(flatdat) #stats
fmax=np.max(flatdat)
fmean=np.mean(flatdat)
fstd=np.std(flatdat)
fmed=np.median(flatdat)
fmode=stats.mode(flatdat)
print(f"The minimum is {fmin}, the maximum is {fmax}, the mean is, {fmean}, the standard deviation is {fstd}, the median is {fmed}, the mode is {fmode}.")

fig, ax = plt.subplots() #using oo method

ax.hist(flatdat, color='lightblue') #using ax to plot
ax.axvline(x=fmean, color='hotpink', label=f'Mean={fmean:.2f}') #using the axvline to plot the ines
ax.axvline(x=fmed, color='purple', label=f'Median={fmed:.2f}')
ax.axvline(x=fmode[0], color='teal', label=f'Mode={fmode[0]}')
ax.legend()
plt.show()

datfloat = np.asarray(dat) #converting teh data to floats
immed=np.median(datfloat) #calculating the median

labelx = hdr['CTYPE1'] #extracting the ctypes
labely = hdr['CTYPE2']
hdr['BUNIT']='Counts'#adding the counts header thingy

fig2, ax2 = plt.subplots() #creating a new plot
im= ax2.imshow(datfloat, origin= 'lower', vmin=immed) #plotting the image data using the median as vmin
ax2.set_xlabel(labelx)#setting the x and y labels
ax2.set_ylabel(labely)
cbar = fig.colorbar(im, ax=ax2)
cbar.set_label(hdr.get('BUNIT')) #setting the units label
plt.show()


newm101= fits.PrimaryHDU(data=datfloat)
newm101.header=hdr
newm101.writeto('m101float.fits', overwrite=True, output_verify='ignore')
```

```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

m101= fits.open('m101.fits')
dat=m101[0].data  

fig, ax = plt.subplots() 

im=ax.imshow(dat, origin='lower',vmax=np.percentile(dat, 50)) #using the percentile to see where there is lots of noise

noise=dat[0:20, 0:20]
noisearr=np.array(noise)#extracting the data

noisearr = noisearr.astype(float)#so that squaring it works correctly

squaredn=noisearr**2 #defining the diff parts of the xrms function
inside=np.mean(squaredn)
xrms=(inside)**(1/2)

print(f"RMS={xrms}")

xmean=np.mean(noise)
xstd=np.std(noise)

print(f"The mean is {xmean} and the standard deviation is {xstd}") #differs because rms tests dev from 0, std from mean
#stdv s better
```

```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

nm0= fits.open('noisemap0.fits') #opening teh files
nm1=fits.open('noisemap1.fits')
nm2=fits.open('noisemap2.fits')

dat0=nm0[0].data #extracting the image data
dat1=nm1[0].data
dat2=nm2[0].data

std0=np.std(dat0) #calculating each std 
std1=np.std(dat1)
std2=np.std(dat2)

print(F"Standard deviation 0: {std0}, standard deviation 1: {std1}, standard deviation 2: {std2}")

stacked=np.stack((dat0, dat1, dat2)) #stacking the images
stackedim=np.mean(stacked, axis=0) #creating a new image based off of the mean of the stacked ones

stdstack=np.std(stackedim) #std of the new image

print(F"Standard deviation in the mean image: {stdstack}")

N=3 #nb of values
stdtot=(np.sqrt((std0**2)+(std1**2)+(std2**2)))/N #formula for the std 

print(F"Standard deviation using the formula: {stdtot}") #almost exactly teh same
```

```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.table import Table


# Open the FITS file
hdul = fits.open('APOGEE_stars.fits')
hdul.info()

# Access the primary data (in this case, it is a Table)
data = hdul[1].data  # Often, the first extension contains the table
colnames = hdul[1].columns.names # List of the column names

#type, dim, format, header locxation are different

mask= ((data['STARFLAG'] == 0) & #creating a mask to filter the data
    (data['ASPCAPFLAG'] == 0) &
    (data['SNR'] > 10) &
    (data['GAIAEDR3_PARALLAX'] / data['GAIAEDR3_PARALLAX_ERROR'] > 5) &
    (~np.isnan(data['GAIAEDR3_PHOT_G_MEAN_MAG'])) &
    (~np.isnan(data['GAIAEDR3_PHOT_BP_MEAN_MAG'])) &
    (~np.isnan(data['GAIAEDR3_PHOT_RP_MEAN_MAG'])))

newdata=[] #making a list of the new data

for k in range (len(data)): #appending 0 and 1 if the conditons are met
    if mask[k]==True:
       newdata.append(1)
    else:
        newdata.append(0)

data=Table(data, names=colnames)#making teh data nto a table

data.add_column(newdata, name='DATAFLAG') #adding a column

print(data)

para=data['GAIAEDR3_PARALLAX'] #parallax
mg=data['GAIAEDR3_PHOT_G_MEAN_MAG']#app mag
Mg=mg+(5*np.log10(para/1000))+5 #abs magnitude

mask2=(data['DATAFLAG']==1) #create new mask 

magdata=[] #making a new list for the abs mag
for j in range (len(data)):
    if mask[j]==True:
        magdata.append(-1.0)
    else:
        magdata.append(Mg[j])

data.add_column(magdata, name='Absolute magitude')

print(data)

data.write('APOGEE_update.fits', overwrite=True)
```

```python
class Fibonacci:
    """Class for calculating Fibonacci sequence"""
    def __init__(self, n, m):   
        self.n=n
        self.m=m
        
    def fibo_n(self):
        a=0 #setting the initial values
        b=1
        for i in range (2, self.n):
            c=a+b
            a=b
            b=c
        return b
    
    def fibo_list(self):
        fiblist=[] #creating a list to add the fibonacci values to 
        a=0 #setting the initial values
        b=1
        l=2 #setting the initial length of the list 
        while l<(self.n-1): #ensuring that the length of the list is less than N
            if a%self.m==0: #ensuring that the fib number is divisible by M
                fiblist.append(a)
            c=a+b #creating the new fib value which is the sum of the previous two
            a=b #updating the 'previous' values
            b=c
            l+=1
        return fiblist
        
    
fib=Fibonacci(100,7)

print(fib.fibo_n())
print(fib.fibo_list())
```
