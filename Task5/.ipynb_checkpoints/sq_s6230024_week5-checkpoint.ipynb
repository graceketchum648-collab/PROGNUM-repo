```python
import numpy as np

data = np.loadtxt('colms.txt', comments="#")
x= data[:,0]; y= data[:,1] #taking first and second columns

minix=np.min(x) #using np functions 
miniy=np.min(y)
maxx=np.max(x)
maxy=np.max(y)
meanx=np.mean(x)
meany=np.mean(y)

y_med=np.median(y)
y2 = (y - y_med)**2 

print(f"The minimum of x is {minix}, the maximum of x is {maxx}, and the mean of x is {meanx}")
print(f"The minimum of y is {miniy}, the maximum of y is {maxy}, and the mean of y is {meany}")
print(y2)
```

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

con=np.concatenate((a, b)) #concatenating the arrays

print(con) #the shape is a list with 1 row of 6 values

a = np.array([[1, 2], [3, 4]]) #an array with rows of 2 values each
b = np.array([[5, 6], [7, 8]])

con1=np.concatenate((a, b))
con2=np.concatenate((a, b), axis=0)
con3=np.concatenate((a, b), axis=1)

print(con1) #4 rows of two values each
print(con2)#no dif
print(con3) #2 rows with 4 columns each
```

```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import numpy as np

im = imageio.imread('m101BW.jpg')

plt.contour(im, levels=np.percentile(im, [99.9]), colors='r')
plt.imshow(im, cmap='jet')
plt.show()
#mapping the 10th percentile

perc=np.percentile(im, 90) #filtering the 90 percentile
#maskim=np.ma.mask_rowcols(im>perc, axis=None) #showing the data bigger than the 90th percentile
maskim=np.where(im>perc, im, perc)
plt.imshow(maskim, cmap='jet')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('ngc6946.txt', comments="#")
resdata=data.reshape(40, 50, 70)

zprof=resdata[:, 25, 35]

#elementary stats
minz=np.min(zprof)
maxz=np.max(zprof)
sumz=np.sum(zprof)
meanz=np.mean(zprof)
stdz=np.std(zprof)


plt.text(25, 0.04, f"Min={minz}")
plt.text(25, 0.045, f"Max={maxz}")
plt.text(25, 0.05, f"Sum={sumz}")
plt.text(25, 0.055, f"Mean={meanz}")
plt.text(15, 0.06, f"Standard dev={stdz}")
plt.plot(zprof)
print(zprof)
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('ngc6946.txt', comments="#")
resdata=data.reshape(40, 50, 70) #reshaping the data
firstplane=resdata[0] #having it be the first plane

meand=np.mean(firstplane) #elementary stast
stdd=np.std(firstplane)

l1=meand #creating the levels
l2=meand+stdd
l3=meand+stdd*2

plt.contour(resdata[0], levels=[l1, l2, l3], colors=['purple','lightblue','pink']) #plotting the levels 

plt.imshow(firstplane, origin='lower')
plt.colorbar()
plt.show()
```

```python
from scipy.constants import c

class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      # Calcium H line in rest frame
    c=c
    
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
    def redshift(self):        
        z = (self.caH - self.caH_0)/self.caH
        return z
    def velocity(self):
        v=(self.redshift()*self.c)/1000
        return v
    
gal1 = Galaxy("NGC1832", 3994.7)

print(f"The velocity of the galaxy is {gal1.velocity()} km/s")
```

```python
import numpy as np
from matplotlib.pyplot import figure, show
import matplotlib.patches as patches
import matplotlib.pyplot as plt

Nx = 400
Ny = 300
data = np.random.random((Ny,Nx))

fig, ax = plt.subplots()
plt.imshow(data, interpolation='none', origin="lower", cmap='gray')

semi_major=100*2
semi_minor=50*2
position_angle=30.0
xc=Nx/2
yc=Ny/2

ellipse=patches.Ellipse((xc, yc), semi_major, semi_minor, angle=position_angle)
ax.add_patch(ellipse)

plt.show()
```
