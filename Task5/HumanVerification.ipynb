## Task 3 - Habitable Zone Check (4pt)

The habitable zones for a star depends on its luminosity and is given as a range in **AU** (astronomical units). Given known stellar radius and surface temperature, the stellar luminostiy can be calculated with *Stefan–Boltzmann law*: 

$$ L = 4\pi R^2 \sigma T^4,
$$ 

where L, R, and T represent the stellar lumninosity, radius, and surface temperature, respectively, and $\sigma$ is the Stefan–Boltzmann constant. 

The habitable zone is discussed relative to our Sun. In solar unit, $\frac{L}{L_{\odot}}=(\frac{R}{R_{\odot}})^2(\frac{T}{T_{\odot}})^4$, and the habitable distance: $$ D_\mathrm{h} = \sqrt{\frac{L}{L_{\odot}}}\ \mathrm{(AU)} $$


```python
# Unit: solar radius
radii = {
    "Sirius": 1.71,
    "Sun": 1,
    "Proxima Centauri": 0.154,
    "Beta Centauri": 8.4

```

The stars listed above, each has one or more planets at different distances to its star as listed below. The habitable zone of a star can be estimated by $$ D_\mathrm{inner} = \sqrt{\frac{D_h^2}{1.06}}$$ $$ D_\mathrm{outer} = \sqrt{\frac{D_h^2}{0.36}}$$


```python
planets = {
    "Sirius": [4.1, 8.2, 5.0],
    "Sun": [0.39, 0.72, 1.0, 1.80],
    "Proxima Centauri": [0.05, 0.12],
    "Beta Centauri": [100.2, 153.5, 250.3, 280.5]
}
```

* Calculate the habitable zone for each star.
* Loop through each star and its planets.
* Check whether each planet is within the habitable zone of their host star.
* Using **f-string** to print your result in the following format for each star.

> Star: Sun is G-type star\
> Planet at 0.39 AU: Not Habitable\
> Planet at 0.72 AU: Not Habitable\
> Planet at 1.0 AU: Habitable\
> Planet at 1.80 AU: Not Habitable


````python
import math

# Star data: Radius (R/R_sun) and Temperature (K)
# Note: T_sun is approx 5778K
stars_data = {
    "Sirius": {"radius": 1.71, "temp": 9940, "type": "A-type"},
    "Sun": {"radius": 1.0, "temp": 5778, "type": "G-type"},
    "Proxima Centauri": {"radius": 0.154, "temp": 3042, "type": "M-type"},
    "Beta Centauri": {"radius": 8.4, "temp": 25000, "type": "B-type"}
}

planets = {
    "Sirius": [4.1, 8.2, 5.0],
    "Sun": [0.39, 0.72, 1.0, 1.80],
    "Proxima Centauri": [0.05, 0.12],
    "Beta Centauri": [100.2, 153.5, 250.3, 280.5]
}

T_sun = 5778

for star, info in stars_data.items():
    # 1. Calculate relative luminosity: (R/R_sun)^2 * (T/T_sun)^4
    l_ratio = (info["radius"]**2) * ((info["temp"] / T_sun)**4)
    
    # 2. Calculate Dh and the Habitable Zone boundaries
    d_h = math.sqrt(l_ratio)
    d_inner = math.sqrt((d_h**2) / 1.06)
    d_outer = math.sqrt((d_h**2) / 0.36)
    
    # 3. Print Header
    print(f"Star: {star} is {info['type']} star")
    
    # 4. Check each planet
    for distance in planets[star]:
        if d_inner <= distance <= d_outer:
            status = "Habitable"
        else:
            status = "Not Habitable"
        
        print(f"Planet at {distance} AU: {status}")
    print("-" * 30) # Separator for readability
````
